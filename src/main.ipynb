```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
model = BertForSequenceClassification.from_pretrained("bert-large-uncased", num_labels=NUM_LABELS, id2label=id2label, label2id=label2id)
model.to(device)
```

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(df_org, test_size=0.3, stratify=df_org['email_class'], random_state=42)

val_df, test_df = train_test_split(temp_df, test_size=0.3333, stratify=temp_df['email_class'], random_state=42)

train_df = train_df.sample(frac=1.0, random_state=42)
val_df = val_df.sample(frac=1.0, random_state=42)
test_df = test_df.sample(frac=1.0, random_state=42)
train_body = list(train_df.body)
val_body = list(val_df.body)
test_body = list(test_df.body)

train_subject = list(train_df.subject)
val_subject = list(val_df.subject)
test_subject = list(test_df.subject)

train_labels = list(train_df.labels)
val_labels = list(val_df.labels)
test_labels = list(test_df.labels)
```

```python

train_text=["Subject:" + subject + ".\n" + body for subject, body in zip(train_subject, train_body)]
val_text=["Subject:" + subject + ".\n" + body for subject, body in zip(val_subject, val_body)]
test_text=["Subject:" + subject + ".\n" + body for subject, body in zip(test_subject, test_body)]
# print(train_text)
```

```python
len(train_text), len(val_text), len(test_text)
```

```python
train_encodings = tokenizer(train_text, truncation=True, padding=True, max_length=512)
val_encodings  = tokenizer(val_text, truncation=True, padding=True, max_length=512)
test_encodings = tokenizer(test_text, truncation=True, padding=True, max_length=512)
```

```python
class DataLoader(Dataset):

    def __init__(self, encodings, labels):
        
        self.encodings = {key: torch.tensor(val) for key, val in encodings.items()}
        self.labels = torch.tensor(labels)

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = self.labels[idx]
        return item

    def __len__(self):
        
        return len(self.labels)
```

```python
train_dataloader = DataLoader(train_encodings, train_labels)
print(train_dataloader[0])

val_dataloader = DataLoader(val_encodings, val_labels)



test_dataset = DataLoader(test_encodings, test_labels)
```

```python
from transformers import TrainingArguments, Trainer
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def compute_metrics(pred):

    labels = pred.label_ids

    preds = pred.predictions.argmax(-1)

    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='macro')

    acc = accuracy_score(labels, preds)

    return {

        'Accuracy': acc,

        'F1': f1,

        'Precision': precision,

        'Recall': recall

    }
```

```python
os.mkdir("../data/multi-class-logs")
os.mkdir("../data/Model")
```

```python
training_args = TrainingArguments(

    output_dir='../data/Model', 

    do_train=True,

    do_eval=True,

    num_train_epochs=10,              

    per_device_train_batch_size=16,  

    per_device_eval_batch_size=32,

    warmup_steps=100,                

    weight_decay=0.01,

    logging_strategy='steps',

    logging_dir='../data/multi-class-logs',            

    logging_steps=50,

    eval_strategy="steps",

    eval_steps=50,

    save_strategy="steps", 

    fp16=True,

    load_best_model_at_end=True

)
```

```python
trainer = Trainer(

    model=model,                       

    args=training_args,                 

    train_dataset=train_dataloader,         

    eval_dataset=val_dataloader,            

    compute_metrics= compute_metrics

)
```

```python
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
os.environ["TORCH_USE_CUDA_DSA"] = "1"
```

```python
trainer.train()
```

```python
q=[trainer.evaluate(eval_dataset=df_org) for df_org in [train_dataloader, val_dataloader, test_dataset]]

pd.DataFrame(q, index=["train","val","test"]).iloc[:,:5]
```

```python
def predict(text):

    inputs = tokenizer(text, padding=True, truncation=True, max_length=512, return_tensors="pt").to("cuda")
    outputs = model(**inputs)

    probs = outputs[0].softmax(1)

    pred_label_idx = probs.argmax()

    pred_label = model.config.id2label[pred_label_idx.item()]
    return probs, pred_label_idx, pred_label
```
